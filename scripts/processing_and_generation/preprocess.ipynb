#Default (Deprecated)

```python
def process_data(set_name):
  columns = ['essay_id', 'full_text',
       'holistic_essay_score', 'source_text', 'prompt_name', 'assignment', 'gender', 'grade_level',
       'ell_status', 'race_ethnicity', 'economically_disadvantaged',
       'student_disability_status']
  types = {'holistic_essay_score': 'object', 'grade_level': 'object'}
  data = pd.read_csv(f'/content/drive/MyDrive/COS597H/data/persuade_corpus_2.0_{set_name}.csv', usecols=columns, dtype=types).fillna("")
  print(data.dtypes)

  examples = []
  unique_ids = np.unique(data['essay_id'])

  for id in unique_ids:
    ex = {}
    subset = data[data['essay_id'] == id]
    for col in columns:
      ex[col] = subset[col].iloc[0]
    examples.append(ex)

  with open(f'/content/drive/MyDrive/COS597H/data/{set_name}.json', 'w') as out_file:
    for example in examples:
      out_file.write(json.dumps(example) + '\n')
```

#Demographic (Indiv Only)

```python
import math

def process_data_demo(set_name, ids):
  columns = ['essay_id', 'full_text',
       'holistic_essay_score', 'prompt_name', 'assignment', 'gender', 'grade_level',
       'ell_status', 'race_ethnicity', 'economically_disadvantaged',
       'student_disability_status']
  types = {'holistic_essay_score': 'object', 'grade_level': 'object'}
  data = pd.read_csv(f'/content/drive/MyDrive/COS597H/data/persuade_corpus_2.0_{set_name}.csv', usecols=columns, dtype=types).fillna("")
  print(data.dtypes)

  examples = []
  cf_examples = []

  subset_size = len(ids)/6
  subset_names = ["gender", "grade_level", "ell_status", "race_ethnicity", "economically_disadvantaged", "student_disability_status"]
  subset_options = [["M", "F"],
                    ["8", "9", "10", "11", "12"],
                    ["Yes", "No"],
                    ["American Indian/Alaskan Native", "Asian/Pacific Islander", "Black/African American",  "Hispanic/Latino", "Two or more races/Other", "White"],
                    ["Economically disadvantaged", "Not economically disadvantaged"],
                    ["Identified as having disability", "Not identified as having disability"]]


  default_fields = ['essay_id', 'full_text', 'holistic_essay_score', 'prompt_name', 'assignment']

  for i, id in enumerate(ids):
    ex = {}
    subset = data[data['essay_id'] == id]
    for col in default_fields:
      ex[col] = subset[col].iloc[0]
    demo_i = math.floor(i / subset_size)
    ex['demographic_variable'] = subset_names[demo_i]
    ex['demographic_value'] = str(subset[subset_names[demo_i]].iloc[0])
    real_val = ex['demographic_value']
    examples.append(ex)
    for option in subset_options[demo_i]:
      if option != real_val:
        copy = ex.copy()
        copy['demographic_value'] = option
        cf_examples.append(copy)

  with open(f'/content/drive/MyDrive/COS597H/data/{set_name}_indiv_demo_shuffled2.json', 'w') as out_file:
    for example in examples:
      out_file.write(json.dumps(example) + '\n')


  with open(f'/content/drive/MyDrive/COS597H/data/{set_name}_indiv_demo_shuffled_cf2.json', 'w') as out_file:
    for example in cf_examples:
      if example['demographic_variable'] == "race_ethnicity":
        print(example['demographic_value'])
      out_file.write(json.dumps(example) + '\n')
```
